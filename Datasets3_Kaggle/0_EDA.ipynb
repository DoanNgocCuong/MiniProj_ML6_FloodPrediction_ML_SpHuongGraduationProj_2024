## 3.3. Phân tích khám phá dữ liệu (Exploratory Data Analysis - EDA)

1. Phân tích ngoại lai (Outlier Analysis)


```d
Danh sách 10 cột quan trọng tự chọn theo cảm nhận để phân tích ngoại lai:

MonsoonIntensity - Lượng mưa: Một yếu tố ảnh hưởng lớn đến nguy cơ lũ lụt.
Deforestation - Độ che phủ rừng: Ảnh hưởng đến khả năng hấp thụ nước và kiểm soát dòng chảy.
Urbanization - Đô thị hóa: Tăng nguy cơ lũ lụt do giảm diện tích thấm nước tự nhiên.
DrainageSystems - Hệ thống thoát nước: Ảnh hưởng trực tiếp đến mức độ nghiêm trọng của lũ.
DamsQuality - Chất lượng đập: Quyết định khả năng kiểm soát lũ.
PopulationScore - Mật độ dân số: Yếu tố quan trọng trong việc đánh giá tác động của lũ.
CoastalVulnerability - Tính dễ tổn thương ven biển: Đặc biệt quan trọng ở khu vực gần biển.
Siltation - Xói mòn: Tác động đến khả năng lưu trữ nước của các con đập.
AgriculturalPractices - Thực hành nông nghiệp: Ảnh hưởng đến dòng chảy bề mặt và xói mòn.
FloodProbability - Xác suất lũ lụt: Biến mục tiêu chính để phân tích các yếu tố liên quan.
```

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Chọn 10 cột quan trọng để phân tích
columns_to_analyze = [
    'MonsoonIntensity', 'Deforestation', 'Urbanization', 'DrainageSystems',
    'DamsQuality', 'PopulationScore', 'CoastalVulnerability', 'Siltation',
    'AgriculturalPractices', 'FloodProbability'
]

# Phát hiện ngoại lai bằng Boxplot
plt.figure(figsize=(15, 12))
for idx, column in enumerate(columns_to_analyze, 1):
    plt.subplot(3, 4, idx)  # Điều chỉnh lưới để hiển thị vừa 10 cột
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot của {column}')
plt.tight_layout()
plt.show()

# Xử lý ngoại lai bằng IQR
rows_before = df.shape[0]  # Số lượng hàng ban đầu
for column in columns_to_analyze:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Lọc ngoại lai cho cột hiện tại
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

rows_after = df.shape[0]  # Số lượng hàng sau khi xử lý
print(f"\n=== Dữ Liệu Sau Khi Loại Bỏ Ngoại Lai ===")
print(f"Số lượng hàng ban đầu: {rows_before}")
print(f"Số lượng hàng sau khi loại bỏ ngoại lai: {rows_after}")
print(f"Số lượng hàng bị loại bỏ: {rows_before - rows_after}")
print(df.describe())
````

2. Phân phối của các biến

````python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# Loại bỏ các cột không cần thiết
columns_to_analyze = df.columns.drop(['id', 'FloodProbability'])
num_columns = len(columns_to_analyze)

# Tính toán số hàng và cột cho lưới subplot
num_rows = math.ceil(num_columns / 4)  # 4 cột mỗi hàng

# Tạo biểu đồ
plt.figure(figsize=(15, num_rows * 4))  # Kích thước phù hợp với số hàng
for idx, column in enumerate(columns_to_analyze, 1):
    plt.subplot(num_rows, 4, idx)  # Tạo subplot tại vị trí tương ứng
    sns.histplot(df[column], kde=True, bins=29)
    plt.title(f'Phân Phối của {column}')
plt.tight_layout()  # Tự động sắp xếp các biểu đồ
plt.show()
````

### 3.3.5. Trực quan hóa kết quả
```
Trực quan hóa các kết quả phân tích để hiểu rõ hơn về dữ liệu và hỗ trợ trong việc xây dựng mô hình.
Biểu đồ xu hướng (Trend Lines):
Vẽ các đường xu hướng để nắm bắt các mô hình dài hạn và biến động ngắn hạn của các yếu tố ảnh hưởng đến lũ lụt. Điều này giúp nhận diện các xu hướng tích cực hoặc tiêu cực trong dữ liệu, từ đó đưa ra các biện pháp phòng chống lũ lụt hiệu quả hơn.
Biểu đồ phân tán (Scatter Plots):
Vẽ biểu đồ phân tán giữa các yếu tố và biến mục tiêu FloodProbability để nhận diện các mối tương quan tiềm năng và các mô hình phi tuyến tính. Điều này giúp hiểu rõ hơn về cách mà các yếu tố ảnh hưởng đến xác suất lũ lụt và đưa ra các quyết định chọn lọc đặc trưng phù hợp.
Biểu đồ ma trận tương quan (Correlation Heatmap):
Trình bày lại mối tương quan giữa các yếu tố bằng biểu đồ heatmap để dễ dàng nhận diện các yếu tố quan trọng và loại bỏ các yếu tố dư thừa. Biểu đồ này giúp nhanh chóng xác định các yếu tố có mối tương quan cao với nhau hoặc với biến mục tiêu, từ đó tối ưu hóa quá trình chọn lọc đặc trưng cho mô hình dự báo.
Biểu đồ hộp (Box Plots):
Sử dụng box plots để so sánh phân phối của các yếu tố giữa các nhóm lũ lụt và không lũ lụt, giúp xác định các yếu tố có ảnh hưởng lớn. Điều này giúp hiểu rõ hơn về sự khác biệt giữa các nhóm dữ liệu và đưa ra các quyết định chọn lọc đặc trưng phù hợp cho mô hình dự báo.
```
