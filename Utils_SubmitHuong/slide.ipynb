### **Slide: Xử Lý Dữ Liệu**

---

### **1. Mục Đích Xử Lý Dữ Liệu**
- Đảm bảo **dữ liệu sạch** và **đầy đủ** cho mô hình học máy.
- Loại bỏ các yếu tố gây **nhiễu** hoặc không phù hợp.
- Chuẩn bị dữ liệu ở định dạng **chuẩn hóa** và tối ưu hóa hiệu suất.

---

### **2. Quy Trình Xử Lý Dữ Liệu**
#### **Bước 1: Làm Sạch Dữ Liệu**
- **Xử lý giá trị thiếu**:
  - Điền giá trị trung bình, trung vị.
  - Loại bỏ các dòng dữ liệu thiếu nếu không thể impute.
- **Xử lý ngoại lai (Outliers)**:
  - Phát hiện bằng **IQR** hoặc **Z-score**.
  - Điều chỉnh hoặc loại bỏ các giá trị bất thường.

---

#### **Bước 2: Chuẩn Hóa Dữ Liệu**
- **Chuẩn hóa (Normalization)**:
  - Dùng **Min-Max Scaler** đưa giá trị về khoảng [0, 1].
  - Đảm bảo các đặc trưng có ảnh hưởng tương đương.
- **Biến đổi (Transformation)**:
  - Sử dụng **log** hoặc **Box-Cox** để làm phẳng phân phối dữ liệu lệch.

---

#### **Bước 3: Trích Xuất và Chọn Lọc Đặc Trưng**
- **Trích xuất đặc trưng (Feature Extraction)**:
  - Tính toán các đặc trưng kỹ thuật: 
    - **Moving Averages**: Đường trung bình động.
    - **Volatility Indicators**: Chỉ số biến động.
  - Tạo đặc trưng tương tác: Ví dụ, **lượng mưa x độ che phủ rừng**.
- **Phân tích đặc trưng (Feature Analysis)**:
  - Loại bỏ đặc trưng không quan trọng bằng **tương quan Pearson** hoặc **PCA**.
  - Giảm chiều dữ liệu nhưng vẫn giữ thông tin quan trọng.

---

#### **Bước 4: Phân Chia Dữ Liệu**
- **Tỷ lệ chia**:
  - Tập huấn luyện: 70%.
  - Tập kiểm tra: 30%.
- **Cross-Validation**:
  - Sử dụng **K-Fold Cross-Validation** để tăng tính chính xác của mô hình.

---

### **3. Kết Quả Xử Lý Dữ Liệu**
- Dữ liệu sạch, chuẩn hóa và sẵn sàng cho việc triển khai mô hình.
- Xác định được các **đặc trưng quan trọng** ảnh hưởng lớn đến nguy cơ lũ lụt.
- Tăng cường độ chính xác và hiệu suất dự báo.
